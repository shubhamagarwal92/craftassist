```python
# Imports
import os
import json
import numpy as np
import glob
```

```python
dir_path = "../../data/dialogue_data/"
```

```python
def get_folder_list(dir_root, folder_name=""):
    path_data = os.path.join(dir_root, folder_name)
    folder_list = glob.glob(os.path.join(path_data, '*'))
    print("Total files in folder:", len(folder_list))
    return folder_list
```

```python
# Data dir 
data_dir_list = get_folder_list(dir_path, folder_name="")
data_dir_list
```

```python
with open(data_dir_list[3]) as readme_file:
    readme_text = readme_file.readlines()
print(readme_text)
```

```python
def json_load_in_mem(data_path):
    json_data = json.load(open(data_path))
    print("Total length:", len(json_data))
    print("Data Type:", type(json_data))
    return json_data

def print_line_by_indx(data_obj, indx):
    print(data_obj[indx][1].keys()) # Data specific
    print('\n')
    print(data_obj[indx])
```

```python
generated_dialogues = json_load_in_mem(data_dir_list[2])
```

```python
dial_indx = 6
```

```python
print_line_by_indx(generated_dialogues, dial_indx)
```
